```python
def Ret_Pregunta05():
    df_5 = pd.read_csv('datasets\Fuentes_Consumo_Energia.csv')
    masc = df_5['Entity'] == 'Europe'
    df5bis = df_5[masc]
    df_5.iloc[[df5bis['Hydro_Generation_TWh'].idxmax()]]
    dfinal= df_5.iloc[[df5bis['Hydro_Generation_TWh'].idxmax()]]
    resultado= int(dfinal['Year'])
    return(resultado)
    
    
```
